```
from google.colab import auth
auth.authenticate_user()
import gspread
from oauth2client.client import GoogleCredentials
gc = gspread.authorize(GoogleCredentials.get_application_default())
table = gc.open_by_key('14yQll4suR7vRRH39QddGfqUysfO1X32zUG90KgqbcK8')

#Считаем лист, на котором представлен свот анализ
Worksheet_Swot = table.worksheet('List_1')


#График для opportunities
Characteristics_opportunities = []
Characteristics_opportunities_power = []
row_count = 4

while True:
  if Worksheet_Swot.cell(row_count, 1).value != '':
    Characteristics_opportunities.append(Worksheet_Swot.cell(row_count, 1).value)
    Characteristics_opportunities_power.append(Worksheet_Swot.cell(row_count, 5).value)
    row_count = row_count + 1
  else:
    break

import matplotlib.pyplot as plt


x_pos = []
y_float = []
x_float = []

print('Имя характеристики ', 'Действия ', 'Сила влияния:', '\n')
for i in range(0, len(Characteristics_opportunities)):
  print(Characteristics_opportunities[i], Characteristics_opportunities_power[i])
  x_pos.append(i)
  x_float.append(Characteristics_opportunities[i])
  y_float.append(Characteristics_opportunities_power[i])
fig=plt.figure(figsize=(30,12), dpi=80)
plt.bar(x_pos, y_float, width=0.5, align='center', alpha=0.5, )
plt.xticks(x_pos,  x_float, fontsize=16)
plt.xlabel('Характеристики', fontsize=40)
plt.ylabel('Мощность воздействия', fontsize=40)
plt.title('opportunities', fontsize=50)
plt.grid(True, color='r', linestyle='-', linewidth=0.8)
plt.draw()
plt.show()
```

```
#Общий график
from google.colab import auth
auth.authenticate_user()
import gspread
from oauth2client.client import GoogleCredentials
gc = gspread.authorize(GoogleCredentials.get_application_default())
table = gc.open_by_key('1ObdM0bKZzgVMnm3ascdUP34zzKRnjXff1n76R7xDJ44')

#Считаем лист, на котором представлен свот анализ
Worksheet_Swot = table.worksheet("Swot_List_1")
#print(Worksheet_Swot)

#Считаем все характеристики и суммируем их силу 
row_count = 4
opportunities = 0.0
while True:
  if Worksheet_Swot.cell(row_count,4 ).value != '':
    opportunities = opportunities + float(Worksheet_Swot.cell(row_count,5 ).value)
    row_count += 1
  else:
    break
row_count = 2
Strengths = 0.0

while True:
  if Worksheet_Swot.cell(row_count, 6).value != '':
    Strengths = Strengths + float(Worksheet_Swot.cell(row_count, 10).value)
    row_count += 1
  else:
    break 

row_count = 2
threats = 0.0

while True:
  if Worksheet_Swot.cell(row_count, 11).value != '':
    threats = threats + float(Worksheet_Swot.cell(row_count, 15).value)
    row_count += 1
  else:
    break

row_count = 2
Weakness = 0.0
 
while True:
  if Worksheet_Swot.cell(row_count, 16).value != '':
    Weakness = Weakness + float(Worksheet_Swot.cell(row_count, 20).value)
    row_count += 1
  else:
    break    

Weakness = -Weakness
threats = -threats

Result = Strengths + opportunities + Weakness + threats

import matplotlib.pyplot as plt


x_pos = [1, 2, 3, 4, 5]
y_float = [Strengths, Weakness, opportunities, threats, Result]
x_float = ['Strength', 'Weaknesses', 'Opportunities', 'Threats', 'Result']
print('Strength=', Strengths, '\n' 'Weaknesses=', Weakness, '\n', 'Opportunities=', opportunities, '\n', 'Threats=', threats, '\n', 'Result:', Result )
fig=plt.figure(figsize=(16,12), dpi=80)
plt.bar(x_pos, y_float, width=0.5, align='center', alpha=0.5, )
plt.xticks(x_pos,  x_float, fontsize=16)
plt.xlabel('Характеристики', fontsize=40)
plt.ylabel('Мощность воздействия', fontsize=40)
plt.title('SWOT Анализ', fontsize=50)
plt.grid(True, color='r', linestyle='-', linewidth=0.8)
#plt.draw()
plt.show()






```
